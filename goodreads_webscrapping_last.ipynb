```python
import requests
from bs4 import BeautifulSoup as bts
import pandas as pd
```

```python
def parseUrl(url):
    result = requests.get(url,headers = {'User-Agent': "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36"})
    soup = bts(result.text, 'html.parser')
    return soup
```

```python
TOTAL_LINK = []

for totalLink in range(1,101):
    TOTAL_LINK.append("https://www.goodreads.com/list/show/1.Best_Books_Ever?page="+str(totalLink))
print(TOTAL_LINK)
```

```python
ALL_PRODUCT_URL = []

for products in TOTAL_LINK[::]:
    html = parseUrl(products)
    
    for link in html.findAll('a', class_="bookTitle"):
        ALL_PRODUCT_URL.append("https://www.goodreads.com"+link["href"])
    print(len(ALL_PRODUCT_URL))
```

```python
error_list = []
columns_features = []


for details in ALL_PRODUCT_URL[::]:
    html = parseUrl(details)
    bookName = html.find('h1', class_="gr-h1 gr-h1--serif")
    author = html.find("a", class_="authorName")
    rating = html.find("span", {"itemprop":"ratingValue"})
    ratingCount = html.find("meta", {"itemprop":"ratingCount"})
    reviewCount = html.find("meta", {"itemprop":"reviewCount"})
    summary = html.find("div", {"id":"descriptionContainer"})
    review = html.findAll('div', class_ = 'reviewText stacked')
    rating_elements = html.findAll("span", {"class": "staticStars notranslate"})
    user = html.findAll("span", {"itemprop": "author"})
    
    if bookName == None:
        error_list.append(details)
        continue
    
    if len(review) == 0:
        columns_features.append([bookName.text, author.text, rating.text, ratingCount.text, reviewCount.text, summary.text])
        continue
    else:
        for i, j, f in zip(review, rating_elements, user):
            if len(i.text) == 0:
                continue
            if len(f.text) == 0:
                continue
            rating_user = j.get('title')
            if rating_user == "it was amazing":
                point = 5
            elif rating_user == "really liked it":
                point = 4
            elif rating_user == "liked it":
                point = 3
            elif rating_user == "it was ok":
                point = 2
            else:
                point = 1
            columns_features.append([bookName.text, author.text, rating.text, ratingCount.text, reviewCount.text, summary.text, i.text.strip(), point, f.text.strip()])
```
